```python
import pandas as pd
from sklearn.model_selection import train_test_split

train_path = './datasets/MRPC/msr_paraphrase_train.txt'
test_path = './datasets/MRPC/msr_paraphrase_test.txt' 
```

```python
def split_mrpc_data(random_state=42):
    raw_train_df = pd.DataFrame(columns=['Quality', 'id1', 'id2', 'string1', 'string2'])
    with open(train_path, 'r') as f:
        lines = f.readlines()
        for line in lines[1:]:
            line = line.strip().split('\t')
            new_row = {'Quality': line[0], 'id1': line[1], 'id2': line[2], 'string1': line[3], 'string2': line[4]}
            raw_train_df.loc[len(raw_train_df)] = new_row
    
    # make train, dev set
    train_df, dev_df = train_test_split(raw_train_df, test_size=0.2, random_state=random_state)
         
    # test dataframe
    test_df = pd.DataFrame(columns=['Quality', 'id1', 'id2', 'string1', 'string2'])
    with open(test_path, 'r') as f:
        lines = f.readlines()
        for line in lines[1:]:
            line = line.strip().split('\t')
            new_row = {'Quality': line[0], 'id1': line[1], 'id2': line[2], 'string1': line[3], 'string2': line[4]}
            test_df.loc[len(test_df)] = new_row

    # save train, dev, test set
    train_df.to_csv("./datasets/MRPC/train.tsv", sep='\t', index=False)
    dev_df.to_csv("./datasets/MRPC/dev.tsv", sep='\t', index=False)
    test_df.to_csv("./datasets/MRPC/test.tsv", sep='\t', index=False)
```
